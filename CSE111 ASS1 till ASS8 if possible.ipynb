```python
#task 1 
user = "HOusE"
up = []
low = []
for char in user :
  if char.isupper() :
    up.append(char)
  else :
    low.append(char)

if len(low) >= len(up) :
  print(user.lower())
else :
  print(user.upper())
```

```python
#Task 2
user = "jhg231j213"
digit_flag = False
letter_flag = False

for char in user :
  
  if char >= "0" and char <= "9" :
    digit_flag = True

  if (char >= "a" and char <= "z") or (char >= "A" and char <= "Z") :
    letter_flag = True

if digit_flag and letter_flag :
  print("Mixed")

elif digit_flag == True :
  print("Number")

else :
  print("Word")
```

```python
#TAsk 3 

up_alpha = ""
alpha = "a"

for i in range(0, 26) :
  up_alpha += alpha.upper()
  alpha = chr(ord(alpha) + 1)

def alpha_dec(string) :
  empty = []
  for n in range(len(string)) :
    for m in up_alpha :
      if string[n] == m :
        empty.append(n)

  result = ""
  for x in range(empty[0] + 1, empty[1]) :
    result += string[x]

  if result == "" :
    return "Blank"
  else : 
    return result
    
string = input()
print(alpha_dec(string))
```

```python
#task 4 

def common(user) :
  lst = [i for i in user.split(", ")]
  user1 = ""
  user2 = ""
  for i in lst[0] : 
    for j in lst[1] :
      if i == j :
        user1 += i
  
  for x in user1 :
    for y in lst[1] :
      if x == y and x not in user2 :
        user2 += x
  
  if user1 == "" and user2 == "" :
    return "Nothing in common"
  else :
    return user1 + user2

user = input()
print(common(user))
```

```python
#task 5


up = ""
low = ""
letter = "a"

for i in range(0, 26) :
  up += letter.upper()
  low += letter.lower()
  letter = chr(ord(letter) + 1)

def password(user) :
    missing = ""
    missing_digi = True
    missing_special = True
    missing_upper = True 
    missing_lower = True

    for x in range(len(user)) :
      if user[x].isdigit() :
        missing_digi = False 
    
    for x in range(len(user)) :
      if user[x] in up :
        missing_upper = False 

    for x in range(len(user)) :
      if user[x] in low :
        missing_lower = False

    for x in range(len(user)) :
      if user[x] == "_" or user[x] == "$" or user[x] == "#" or user[x] == "@" :
        missing_special = False

    
    if missing_digi :
      missing += "Digit missing"

    if missing_special :
      missing += "Special Character missing"

    if missing_upper :
      missing += "Uppercase missing"

    if missing_lower :
      missing += "Lowercase missing"


    if missing == "" :
      return "Ok"
    else :
      return missing.rstrip(", ")

user = input()
print(password(user))
```
